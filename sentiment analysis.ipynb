```python
df_tweet = pd.DataFrame(data=df)
```

```python
for index, row in df_tweet.iterrows():
    tweet = row["text"]
    text = TextBlob(tweet)
    subjectivity = text.sentiment.subjectivity #0 to 1
    polarity = text.sentiment.polarity #-1 to +1
    #print (tweet, subjectivity, polarity)
    df_tweet.loc[index, 'subjectivity'] = subjectivity
    df_tweet.loc[index, 'polarity'] = polarity
```
